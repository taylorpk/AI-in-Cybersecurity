```python
import pandas as pd
```

```python
#list of products released from OpenAI by year
```

```python
product_list = pd.read_csv('attack_list_AI_Products.csv')
product_list
```

```python
#list of cyberattacks by year
```

```python
attack_type = pd.read_csv('attack_list - updated_attack_type.csv')
attack_type
```

```python
#Pandas merge
```

```python
attack_merge = pd.merge(product_list, attack_type, on='Year')
attack_merge
```

```python
#creates csv of merged pandas cell above
```

```python
attack_df = attack_merge.to_csv('merged_attack.csv', index=False)
attack_df
```
